```r
library(hise)
library(H5weaver)
```

```r
downloadFileFromProjectFolder('rds', 'AIFI-2021-06-26T00:07:00.713108254Z/Hao-2021_PBMC-Multimodal-Reference_counts.h5')
```

```r
h5 <- h5dump("AIFI-2021-06-26T00:07:00.713108254Z//Hao-2021_PBMC-Multimodal-Reference_counts.h5")
```

```r
for(i in 1:5) {
    set.seed(i)
    message("sampling")
    bcs <- sample(h5$matrix$barcodes, 50)
    message("subsetting")
    sub_h5 <- subset_h5_list_by_observations(h5, bcs)
    message("writing")
    write_h5_list(sub_h5, paste0("a_data_kind/some_cells_",i,".h5"))
}
```

```r
meta <- as.data.frame(h5$matrix$observations)
```

```r
for(i in 1:5) {
    set.seed(i)
    message("sampling")
    sub_meta <- meta[sample(1:nrow(meta), 50),]
    message("writing")
    write.csv(sub_meta, paste0("more_different_data/not_really_flow_",i,".csv"),
              row.names = FALSE, quote = FALSE)
}
```
